```python
from sklearn.cluster import KMeans
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.scatter(data.Age,data['BloodPressure'])
plt.xlabel('Age')
plt.ylabel('BloodPressure')
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(data[['Age','BloodPressure']])
y_predicted
```

```python
data['cluster']=y_predicted
data.head()
```

```python
km.cluster_centers_


df1 = data[data.cluster==0]
df2 = data[data.cluster==1]
df3 = data[data.cluster==2]
plt.scatter(df1.Age,df1['BloodPressure'],color='lightgreen')
plt.scatter(df2.Age,df2['BloodPressure'],color='orange')
plt.scatter(df3.Age,df3['BloodPressure'],color='skyblue')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('Age')
plt.ylabel('BloodPressure}')
plt.legend()
```

#Elbow Method

```python
p = data.iloc[:,[8,3]].values
```

```python
from sklearn.cluster import KMeans
wcss_list=[]
```

```python
for i in range(1,16):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit_predict(p)
    wcss_list.append(kmeans.inertia_)
    
```

```python
wcss_list
```

```python
plt.plot(range(1,16),wcss_list)
plt.title('The Elbow Method Graph')
plt.xlabel('Number of clusters(k)')
plt.ylabel('wcss_list')
plt.show()
```
