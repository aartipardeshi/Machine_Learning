```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = np.array(x).reshape(-1,1)
```

```python
X
```

```python
Y = np.array(y).reshape(-1,1)
```

```python
Y
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,random_state = 0,train_size = 0.70) 
```

```python
from sklearn import linear_model
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_predicted = model.predict(x_test)
```

```python
y_predicted
```

```python
x_test
```

```python
from sklearn.metrics import mean_squared_error
```

```python
rmse = mean_squared_error(y_test,y_predicted)
```

```python
rmse
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error  = mean_absolute_error(y_test,y_predicted)
```

```python
mean_absolute_error
```

```python
from sklearn.metrics import mean_absolute_percentage_error
```

```python
mean_absolute_percentage_error = mean_absolute_percentage_error(y_test,y_predicted)
```

```python
mean_absolute_percentage_error
```

```python
accuracy = 100 - mean_absolute_percentage_error
```
